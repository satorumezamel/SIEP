```python
import pandas as pd
import numpy as np
from datetime import timedelta



def load_eeg_data(file_path):
    df = pd.read_csv(file_path, parse_dates=['TimeStamp'])
    df['TimeStamp'] = df['TimeStamp'].dt.tz_localize('Asia/Tokyo')
    return df

def extract_eeg_data(df, task):
    start_time = pd.Timestamp(task['start']).tz_convert('Asia/Tokyo') - timedelta(seconds=1)
    end_time = pd.Timestamp(task['end']).tz_convert('Asia/Tokyo') +timedelta(seconds=1)
    return df[(df['TimeStamp'] >= start_time) & (df['TimeStamp'] <= end_time)]

def clean_sensor_data_relaxed(df, cols):
    cleaned_data = df[cols].dropna()
    for col in cols:
        mean, std = cleaned_data[col].mean(), cleaned_data[col].std()
        cleaned_data = cleaned_data[(cleaned_data[col] >= mean - 4*std) & (cleaned_data[col] <= mean + 4*std)]
    return cleaned_data

def compute_freq_stats(df):
    stats = {}
    freq_bands = {
        'Delta': ['Delta_AF7', 'Delta_AF8', 'Delta_TP9', 'Delta_TP10'],
        'Theta': ['Theta_AF7', 'Theta_AF8', 'Theta_TP9', 'Theta_TP10'],
        'Alpha': ['Alpha_AF7', 'Alpha_AF8', 'Alpha_TP9', 'Alpha_TP10'],
        'Beta': ['Beta_AF7', 'Beta_AF8', 'Beta_TP9', 'Beta_TP10'],
        'Gamma': ['Gamma_AF7', 'Gamma_AF8', 'Gamma_TP9', 'Gamma_TP10']
    }

    for freq, cols in freq_bands.items():
        cleaned_data = clean_sensor_data_relaxed(df, cols)
        combined_data = cleaned_data.values.flatten()
        combined_data = combined_data[~np.isnan(combined_data)]

        # データが空の場合の処理
        if combined_data.size == 0:
            stats[f'{freq}_mean'] = np.nan
            stats[f'{freq}_peak'] = np.nan
            stats[f'{freq}_stability'] = np.nan
        else:
            stats[f'{freq}_mean'] = combined_data.mean()
            stats[f'{freq}_peak'] = combined_data.max()
            stats[f'{freq}_stability'] = combined_data.std()

    return stats
    

all_stats = []

data = pd.read_csv(eeg_file_path)


df_eeg = load_eeg_data(eeg_file_path)

sensor_cols = [
    'Delta_AF7','Delta_AF8','Delta_TP9','Delta_TP10',
    'Theta_AF7','Theta_AF8','Theta_TP9','Theta_TP10',
    'Alpha_AF7','Alpha_AF8','Alpha_TP9','Alpha_TP10',
    'Beta_AF7','Beta_AF8','Beta_TP9','Beta_TP10',
    'Gamma_AF7','Gamma_AF8','Gamma_TP9','Gamma_TP10'
]

for task in quantum_tasks:
    eeg_data = extract_eeg_data(df_eeg, task)
    if eeg_data.empty:
        print(f"Warning: No data for task {task['task_id']}")
        continue
    stats = compute_freq_stats(eeg_data)
    all_stats.append(stats)



eeg_stats = pd.DataFrame(all_stats).round(3)


eeg_stats_dict = {col: eeg_stats[col].tolist() for col in eeg_stats.columns}


print("eeg_stats = pd.DataFrame({")
for i, (key, val) in enumerate(eeg_stats_dict.items()):
    comma = ',' if i < len(eeg_stats_dict)-1 else ''
    print(f"    '{key}': {val}{comma}")
print("})")
```

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr, kendalltau


selected_metrics = 'all'  #  'mean', 'peak', 'stability', 'all' 



if selected_metrics != 'all':
    selected_features = [col for col in eeg_stats.columns if selected_metrics in col]
else:
    selected_features = eeg_stats.columns


results = []
for q_state in quantum_counts.columns:
    for eeg_feat in selected_features:
        clean_data = pd.concat([quantum_counts[q_state], eeg_stats[eeg_feat]], axis=1).dropna()
        if clean_data.empty or len(clean_data) < 2:
            continue  # データ数が不足した場合はスキップ
        
        pearson_r, pearson_p = pearsonr(clean_data.iloc[:, 0], clean_data.iloc[:, 1])
        spearman_r, spearman_p = spearmanr(clean_data.iloc[:, 0], clean_data.iloc[:, 1])
        kendall_r, kendall_p = kendalltau(clean_data.iloc[:, 0], clean_data.iloc[:, 1])

        results.append({
            'Quantum_State': q_state,
            'EEG_Feature': eeg_feat,
            'Pearson_r': pearson_r,
            'Pearson_p': pearson_p,
            'Spearman_r': spearman_r,
            'Spearman_p': spearman_p,
            'Kendall_r': kendall_r,
            'Kendall_p': kendall_p
        })

results_df = pd.DataFrame(results)


significant = results_df[
    (results_df['Pearson_p'] < 0.05) |
    (results_df['Spearman_p'] < 0.05) |
    (results_df['Kendall_p'] < 0.05)
]


significant[['Quantum_State', 'EEG_Feature', 
                   'Pearson_r', 'Pearson_p',
                   'Spearman_r', 'Spearman_p',
                   'Kendall_r', 'Kendall_p']]
```

```python
from statsmodels.stats.multitest import multipletests


pre_filtered = results_df[
    (results_df['Pearson_p'] < 0.05) |
    (results_df['Spearman_p'] < 0.05) |
    (results_df['Kendall_p'] < 0.05)
].copy()


for method in ['Pearson', 'Spearman', 'Kendall']:
    pvals = pre_filtered[f'{method}_p']
    _, adj_pvals = multipletests(pvals, method='fdr_bh')[:2]
    pre_filtered[f'{method}_p_adj'] = adj_pvals


significant = pre_filtered[
    (pre_filtered['Pearson_p_adj'] < 0.05) |
    (pre_filtered['Spearman_p_adj'] < 0.05) |
    (pre_filtered['Kendall_p_adj'] < 0.05)
]


significant[['Quantum_State', 'EEG_Feature', 
             'Pearson_r', 'Pearson_p_adj',
             'Spearman_r', 'Spearman_p_adj',
             'Kendall_r', 'Kendall_p_adj']]
```
